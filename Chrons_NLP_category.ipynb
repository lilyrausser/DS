# ***NLP Selection***

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Tokenize data:

```python
#tokenize categorical data (diagnosis code notes)
df['tokenized_symptoms'] = df['diagnosis_1_category'].apply(lambda x: x.split())
```

Feature Engineering:

```python
#create new data frame to be used in the model; only displaying possible feature options
X = df[['tokenized_symptoms', 'source', 'cd_flag', 'claim_type', 'date_of_service', 'patient_birth_year', 'patient_gender', 'patient_region', 'patient_state', 'speciality', 'npi', 'primary_payer_pay_type', 'total_claim_charge_amount_vendor', 'uc_flag', 'units_of_service', 'new_diagnosed_patient_flag', 'age', 'national_drug_code', 'patient_zip3']]
```

```python
#updating datatypes of the features to match the tokenized data: new table must be created with features as strings
X['source'] = X['source'].astype(str)
X['claim_type'] = X['claim_type'].astype(str)
X['cd_flag'] = X['cd_flag'].astype(str)
X['date_of_service'] = X['date_of_service'].astype(str)
X['patient_birth_year'] = X['patient_birth_year'].astype(str)
X['patient_gender'] = X['patient_gender'].astype(str)
X['patient_region'] = X['patient_region'].astype(str)
X['patient_state'] = X['patient_state'].astype(str)
X['speciality'] = X['speciality'].astype(str)
X['npi'] = X['npi'].astype(str)
X['primary_payer_type'] = X['primary_payer_pay_type'].astype(str)
X['total_claim_charge_amount_vendor'] = X['total_claim_charge_amount_vendor'].astype(str)
X['uc_flag'] = X['uc_flag'].astype(str)
X['units_of_service'] = X['units_of_service'].astype(str)
X['new_diagnosed_patient_flag'] = X['new_diagnosed_patient_flag'].astype(str)
X['age'] = X['age'].astype(str)
X['national_drug_code'] = X['national_drug_code'].astype(str)
```

Testing Accuracy:

```python
#create new table using all of the converted string features from the above cell

#X['features'] = X.apply(lambda row: ' '.join(row['tokenized_symptoms']) + ' ' + row['source'] + ' ' + row['cd_flag'] + ' ' + row['claim_type'] + ' ' + row['date_of_service'] + ' ' + row['patient_birth_year'] + row['patient_gender'] + ' ' + row['patient_region'] + ' ' + row['patient_state'] + ' ' + row['speciality'] + row['npi'] + ' ' + row['primary_payer_pay_type'] + ' ' + row['total_claim_charge_amount_vendor'] + ' ' + row['uc_flag'] + ' '+ row['units_of_service'] + ' ' + row['new_diagnosed_patient_flag'] + ' ' + row['age'], axis = 1)
#X['tokenized_symptoms'] = sample_df['tokenized_symptoms']
X['features'] = X.apply(lambda row: ' '.join(row['tokenized_symptoms']) +' ' + row['cd_flag'] + ' ' + row['claim_type'] + ' ' + row['date_of_service'] + ' ' + row['patient_gender'] + ' ' + row['patient_region'] + ' ' + row['speciality'] + ' ' + row['npi']  + ' ' + row['units_of_service']  + ' ' + row['age'], axis=1)

#define target variable (medication)
y = df['ndc_nonproprietaryname']

#data split
X_train, X_test, y_train, y_test = train_test_split(X['features'], y, test_size = 0.2, random_state = 42)

#vectorize using TF-IDF
vectorizer = TfidfVectorizer()
X_train_features = vectorizer.fit_transform(X_train)
X_test_features = vectorizer.transform(X_test)

#model training
classifier = LogisticRegression(multi_class = 'multinomial', solver = 'lbfgs')
classifier.fit(X_train_features, y_train)

#predicition
predictions = classifier.predict(X_test_features)

#accuracy
accuracy = accuracy_score(y_test, predictions)
print("Accuracy: ", accuracy)
```
