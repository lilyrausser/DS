***What is NCAA's role in athletic funding?***

> The **NCAA** **(National Collegiate Athletic Association)** is the largest Collegiate athletic association in the country. Every year, the sports that the NCAA sponsors train in order to qualify for their sport's NCAA tournament. With the goal in mind of medaling at these prestigeious sporting events, funding becomes a critical component in meeting that goal. While each University's athletics team that participates in an NCAA tournament has their own athletics budget, the NCAA completely funds the large (qualify-basis) tournament at the end of each team's season. While TitleIX was created 50 years ago, the gender funding gap is still not where it needs to be. In order to analyze this further, we will be looking at NCAA reported data from the 2018-2019 season.



***Goal:***



> Due to the fact that data is limited, the only way to understand the funding bias in Collegiate athletics on the basis of gender and race is to split the analysis of the two features in order to clearly see the intersectionality.



**Average Spending Difference Between Male and Female Atheletes in College Athletics**





```python
avg_spending = {'Gender': ['men', 'women'], 'spending': [4285, 2588]}
avg_spending = pd.DataFrame(avg_spending)
# spending_for_lax = spending_for_lax.reset_index().rename_axis('index')

# spending_for_lax.plot(kind = 'pie', y = 'Gender')
avg_spending.plot(kind='pie', labels=avg_spending['Gender'], y='spending', autopct='%1.0f%%')
plt.title("NCAA Average Spending by Gender");
```

**Results:**

> While the NCAA allocates (on average) 4285 dollars per male athlete, they only provide 2588 dollars per female athlete for their respective NCAA tournament; showcasing an almost 2000 dollar difference between the funding available for male and female athletes at the biggest competition of their careers (ESPN).


> From a USAToday Analysis, it was found that, "For every dollar that schools spend on travel, equiptment and recruiting for men's teams, they spent just 71 cents on women."


> Over two seasons, the 29 cent delta adds up to over 125 million spent on men compared to women(NBC).











**Data Visualization Component:**



> A pie chart is chosen to represent this data in order to display the distinct differences between the male and female collegiate athletic tournament funding. A limitation of using a pie chart is that people often have difficulty comparing areas (compared to height with bar or line graph), however, because there is a distinct difference between the two variables, the pie chart is a good option.





***NCAA allocation of resources for Men's, Women's and Coed Collegiate Teams***

```python
AOR = pd.read_excel("Allocation of Resources based on gender.xlsx")
AOR.groupby(['Gender', 'Category']).agg('first')
```

```python
pivot = AOR.pivot_table(index='Category', columns='Gender', values='Percentage', aggfunc=sum)
pivot.plot(kind='bar', figsize=(8,6))


plt.xlabel('Allocation')
plt.ylabel('Resources %')
plt.title('Allocation of Resources by Gender')
plt.legend(title='Gender Sports')


plt.show()
```

**Context:**


> This graph displays the difference in funding for male athletics compared the female athletics within the categories of coach funding, recruiting funding and scholarship funding.



**Results**

> It is important to note that this data is being represented through percentages because the NCAA report was not willing to provide concrete numbers. A statement in an ESPN analysis states, "Direct comparisons between men's and women's championships couldn't be done, because the NCAA does not maintain its records of expenses, ticket sales, amenities, or other items in a standardized manner that would permit such analysis". With that in mind, we can see from the graph that there is on average a significantly higher percentage allocated towards Men's athletics compared to Women's athletics.

# **Funding Bias in College Athletics on the Basis of Race**


---







***History of Racial Bias in Sports***

**The New Plantation**, a book created by Billy Hawkins, follows the
relationship between historically white colleges and Black athletes.
Hawkins believes the complex structure of large collegiate sports creates a plantation mentality that, "has long exploited Black people in the US for economic gain." (Hawkins, 2010)

American racial capitalism informed the structure of higher education since the founding colleges and universities were created; dating back to as early as the 1600s. Due to racial segregation, historically Black universities (HBCU's) allowed for the academic and athletic success for Black people.

Although these higher institutions existed for Black people, funding significantly lacked. Due to the aspect that vast amounts of funding was provided from the state to these historically white institutions (HWI), HWI's could entice Black student athletes from their respective instutitions with the promise of full scholarships and better training facilities. Hawkins, in **The New Plantation**, discusses how this poaching system is exploitative; reminiscant of the plantation system where HWI's 'bought' Black athletes in order to boost their athletic success (Hawkins, 2010). As this process continued, Hawkins described the HWI's exploitative tendencies through their restriction of these poached athlete's ability to monetize their labor. These HWI's felt that they were, "protecting them [Black Athletes] from exploitation"(Hawkins, 2010).

These "plantation dynamics" (as Hawkins puts it) allowed large collegiate athletic programs to profit off of minority athletes for decades. In the 2018-2019 season alone, the large conferences (that obtain teams that compete at NCAA tournaments): ACC, SEC, Big Ten, Pac 12 and Big 12 made 8.3 billion dollars (Gaurdian, 2021).

***The Introduction to the NIL***


---

The NIL was introduced in 2021 in order to limit the plantation dynamics within college athletics (Guardian, 2021). The NIL allows athletes to profit off of their name, image and likeness; something that HWI denied Black and Brown athletes for decades. While the idea of NIL seems to be a great response to the exploitative nature of college athletics, many still believe that the system [of college athletics] needs to be restructured before acts like NIL can truly work.

Many professional athletes are weary with the introduction of NIL in the context of supporting Black and Brown athletes. A former NBA all-star has the opinion that, "Even with [NIL], the same mechanisms of control are still in place, meaning the system is still set up to benefit the players last, not first"(Guardian, 2021). With the context that venerated professional athletes

A current SEC football player states, "It's frustrating for me because NIL doesn't change the fact that I show up eavery Saturday and play in front of thousands of screaming fans and everyone else gets paid"(Guardian, 2021).

# The Intersectionality between Racial and Gendered Bias in Collegiate Athletic Funding


---



This report shows that in order to have women's sports take off the way men's sports had many decades ago, a large investment similar to the one provided to male [HWI] athletics in the 1960s will be needed. Since this was an investment that female and black student athlets [HBCUs] were not allowed to receive, HBCU institutions and female athletics do not have access to the same financial success that HWI male athletics have today.

While the NIL is a step in the right direction it still has its flaws. John Rawls, the author of "Theory of Justice", created the theory of distributive justice, adapting the social contract for 20th century liberalism. While he advocates for the liberal framing of justice centered around the individual, this positioning doesn't account for the racial bias that is deeply embedded in our society. While the NIL tries to address this issue by compensating athletes for the skills and talents, it doesn't address the very prevalent issues associated with lack of equal pay/funding for HBCU/Black athletes and women. When creators of the NIL are blind to the issues of race and gender, one can really see it play out into how sports are funded. NIL ultimately plays under a veil of ignorance, that same ignorance that is prevalent in John Rawl's "Theory of Justice".


The solution to this problem goes beyond a distributive model; paying an equal sum to each sector [HWI institutions, female athletics, HBCUs] cannot make up for the historical delta of billions of dollars in investment. While there isn't a lot of data provided for female and HBCU athletics spending, one can see the intersectionality and the reasoning as to why these areas of athletics weren't able to take off in the same way that male athletics were able to. If more data was provided, the intersectionality would be very clear.

As a result, restorative justice can help close the gap of financial disparities for female and Black [HBCU] athletes. Allowing for a larger investment into these two sectors of athletics will allow for an equal playing field for all athletes, no matter one's uncontrollable features of gender and race.
